```python
file = r"C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_1\Data\Part1.xlsx"

xl1= pd.ExcelFile(file)      # Load spreadsheet
df1 = xl1.parse('Sheet1')    # Load a sheet into a DataFrame by name: df1

df1.dropna(subset = ["participant._round_number"], inplace=True)  # If column "" contains NaN, remove that row

df1.drop(['participant.label','participant._is_bot','participant._index_in_pages',
          'participant._max_page_index','participant._current_app_name','participant._current_page_name',
          'participant._current_page_name','participant.ip_address', 'participant._round_number',
          'participant.exclude_from_data_analysis','participant.visited','participant.mturk_worker_id',
          'participant.mturk_assignment_id','player.id_in_group','player.language','player.payoff',
          'group.id_in_subsession','subsession.round_number', 'session.code', 'session.is_demo'], inplace=True, axis=1)

df1 = df1.loc[:, (df1 != 0).any(axis=0)]
df1c = df1.dropna(how='all', axis=1)
lenght=len(df1c)
df1c.index  = np.arange(lenght)
```
