```python
import pandas as pd
import numpy as np
```

```python
file = r"C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_1\Data\Part2.xlsx"

xl2= pd.ExcelFile(file)      # Load spreadsheet
df2 = xl2.parse('Sheet1')    # Load a sheet into a DataFrame by name: df1

df2.dropna(subset = ["participant._round_number"], inplace=True)  # If column "" contains NaN, remove that row

#number of unique participants
N = df2["participant.code"].nunique()
```

```python
df2.drop(['participant.label','participant._is_bot','participant._index_in_pages',
          'participant._max_page_index','participant._current_app_name','participant._current_page_name',
          'participant._current_page_name','participant.ip_address', 'participant._round_number',
          'participant.exclude_from_data_analysis','participant.visited','participant.mturk_worker_id',
          'participant.mturk_assignment_id','player.id_in_group','player.language','player.payoff',
          'group.id_in_subsession','subsession.round_number', 'session.code', 'session.is_demo'], inplace=True, axis=1)

df2 = df2.dropna(how='all', axis=1)
df2 = df2.loc[:, (df2 != 0).any(axis=0)]
```

```python
a=[]
a1=[]
i=1
for i in [2,3,4,5,6]:
    a=['%d_policy_choice' %i, '%d_Price_UR_A' %i,'%d_Price_UN_A' %i,'%d_Price_O_A' %i,'%d_EV_A' %i,'%d_Revenue_A' %i,'%d_Price_UR_B' %i,
       '%d_Price_UN_B' %i,'%d_Price_O_B' %i,'%d_EV_B' %i,'%d_Revenue_B' %i]
    a1=a1+a
    
df2[a1] = pd.DataFrame(np.empty((1704,55)), index=df2.index)

df2.rename(columns = {'player.policy_choice':'1_policy_choice','player.Price_UR_A':'1_Price_UR_A', 'player.Price_UN_A':'1_Price_UN_A',
                      'player.Price_O_A':'1_Price_O_A','player.EV_A':'1_EV_A','player.Revenue_A':'1_Revenue_A',
                      'player.Price_UR_B':'1_Price_UR_B','player.Price_UN_B':'1_Price_UN_B','player.Price_O_B':'1_Price_O_B',
                      'player.EV_B':'1_EV_B','player.Revenue_B':'1_Revenue_B'}, inplace = True)
```

```python
for i in [1,2,3,4,5]:
    df2['%d_policy_choice' %(i+1)] = np.append(df2['1_policy_choice'][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Price_UR_A' %(i+1)] = np.append(df2['1_Price_UR_A'][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Price_UN_A' %(i+1)] = np.append(df2['1_Price_UN_A'][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Price_O_A' %(i+1)] = np.append(df2['1_Price_O_A' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_EV_A' %(i+1)] = np.append(df2['1_EV_A' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Revenue_A' %(i+1)] = np.append(df2['1_Revenue_A' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Price_UR_B' %(i+1)] = np.append(df2['1_Price_UR_B' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Price_UN_B' %(i+1)] = np.append(df2['1_Price_UN_B' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Price_O_B' %(i+1)] = np.append(df2['1_Price_O_B' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_EV_B' %(i+1)] = np.append(df2['1_EV_B' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
    df2['%d_Revenue_B' %(i+1)] = np.append(df2['1_Revenue_B' ][(N*i):(N*(i+1))].tolist(),np.zeros(1420))
```
