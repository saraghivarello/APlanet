```python
import pandas as pd
import numpy as np
import math
```

## Import data

```python
file1 = r'C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_1\Rearrange_datasets\part1_text.xlsx'

xl1 = pd.ExcelFile(file1)      # Load spreadsheet
df1 = xl1.parse('Sheet1')    # Load a sheet into a DataFrame by name: df1
df1.index  = np.arange(len(df1))
```

```python
file2 = r'C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_1\Rearrange_datasets\part2_text.xlsx'

xl2 = pd.ExcelFile(file2)      # Load spreadsheet
df2 = xl2.parse('Sheet1')    # Load a sheet into a DataFrame by name: df1
df2.drop(['participant.time_started','participant.code','participant.id_in_session','mod'], inplace=True, axis=1)
df2.index  = np.arange(len(df2))
```

```python
file3 = r'C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_1\Rearrange_datasets\part3_text.xlsx'

xl3= pd.ExcelFile(file3)      # Load spreadsheet
df3 = xl3.parse('Sheet1')    # Load a sheet into a DataFrame by name: df1
df3.drop(['participant.time_started','participant.code','participant.id_in_session','mod'], inplace=True, axis=1)
df3.index  = np.arange(len(df3))
```

## Concatenate the three parts

```python
result = pd.concat([df1,df2,df3], axis=1)
result.drop(['Unnamed: 0'], inplace=True, axis=1)
```

```python
#discard rows in result dataframe that have more that 10 nan values
result = result.replace('-', np.nan)
result = result.dropna(thresh=10)
result.reset_index(drop=True, inplace=True)

result.to_excel(r'C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_1\Rearrange_datasets\all_parts.xlsx') 
```
