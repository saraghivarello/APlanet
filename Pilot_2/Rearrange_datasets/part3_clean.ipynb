```python
import pandas as pd
import numpy as np
import math
import matplotlib as plt
```

## Import data

```python
file = r"C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_2\Data\Part3.csv"
df1 = pd.read_csv(file)

df1.drop(['participant._max_page_index','participant._current_app_name','participant._current_page_name', 'participant._index_in_pages',
          'participant._current_page_name','participant.time_started_utc','participant.visited'], inplace=True, axis=1)

#df1 = df1.loc[:, (df1 != 0).any(axis=0)]
df1 = df1.dropna(how='all', axis=1)

#drop rows with  participant codes  'q836peeq','yu68b1d9'
df1 = df1[df1['participant.code'] != 'q836peeq']
df1 = df1[df1['participant.code'] != 'yu68b1d9']
#df1.dropna(subset=['player.SN_car_A'], inplace=True)
lenght=len(df1)
df1.index  = np.arange(lenght)
```

```python
# Pivot the dataframe to have unique row for each participant.code
df1 = df1.pivot_table(index=['participant.code'], aggfunc='first')
df1.reset_index(inplace=True)
```

```python
df1.dropna(subset=['player.SN_car_A'], inplace=True)
df1.reset_index(inplace=True,drop=True)
np.shape(df1)
```

```python
lenght=len(df1)
```

```python
# player.SN_car_A

v=[]
for i in range(lenght):
    if math.isnan(df1['player.SN_car_A'][i]) is True:
        v.append('-')
    elif df1['player.SN_car_A'][i]==0: 
        v.append('A: Public Transport, cycling and walking')
    elif df1['player.SN_car_A'][i]==1: 
        v.append('A: Private vehicle')
df1.insert(loc=df1.columns.get_loc('player.SN_car_A') + 1, column='Considering the following transport mode group:', value=v)
```

```python
answers_percentage = [[1,"0%"],
                          [2,"10%"],
                          [3,"20%"],
                          [4,"30%"],
                          [5,"40%"],
                          [6,"50%"],
                          [7,"60%"],
                          [8,"70%"],
                          [9,"80%"],
                          [10,"90%"],
                          [11,"100%"]]

percentage_dict = {item[0]: item[1] for item in answers_percentage}
#make another dict where values are in inverse order but the keys remain the same
percentage_dict_inverse = {key: value for key, value in zip(percentage_dict.keys(), reversed(percentage_dict.values()))}
```

```python
# EE
column = 'player.EE'
v=[]
for i in range(lenght):
    if math.isnan(df1['player.SN_car_A'][i]) is True:
        v.append('-')
    elif df1['player.SN_car_A'][i]==1: 
        v.append(percentage_dict.get(int(df1[column][i]), '-'))
    elif df1['player.SN_car_A'][i]==0: 
        v.append(percentage_dict_inverse.get(int(df1[column][i]), '-'))
df1.insert(loc=df1.columns.get_loc(column) + 1, column='What share (in %) of other participants in the survey do you think have chosen one of the transport modes in A for their daily trip?', value=v)
```
