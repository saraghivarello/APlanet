```python
# Assuming df1 is your original DataFrame
# Pivot the dataframe to create new columns for each policy choice and other player attributes
df_pivot = df1.pivot(index='participant.code', columns='subsession.round_number', 
                     values=['control_question1', 'control_question2',
                             'attributes_binary_1', 'attributes_binary_2', 'attributes_binary_3',
                             'attributes_binary_4', 'attributes_binary_5', 
                             'alternative1', 'policy_choice', 'Price_UR_A',
                             'Price_UN_A', 'Price_O_A', 'EV_A',
                             'Revenue_A', 'Price_UR_B', 'Price_UN_B',
                             'Price_O_B', 'EV_B', 'Revenue_B'])

# Flatten the multi-level columns and add the prefix
df_pivot.columns = [f'{int(col[1])}_{col[0]}' for col in df_pivot.columns]

# Reset the index to make 'participant.code' a column again
df_pivot.reset_index(inplace=True)

# Drop columns that are all NaN
df_pivot = df_pivot.dropna(axis=1, how='all')

df_pivot.rename(columns={'1_control_question1':'control_question1','1_control_question2':'control_question2','1_alternative1':'alternative1',
                            '8_attributes_binary_1':'attributes_binary_1', '8_attributes_binary_2':'attributes_binary_2',
                                '8_attributes_binary_3':'attributes_binary_3', '8_attributes_binary_4':'attributes_binary_4',
                                        '8_attributes_binary_5':'attributes_binary_5'}, inplace=True)
# Display the pivoted dataframe
df_pivot
```

```python
#drop rows that have nan in column control_question1
df_pivot = df_pivot.dropna(subset=['control_question1'])
df_pivot.reset_index(inplace=True, drop=True)
```
