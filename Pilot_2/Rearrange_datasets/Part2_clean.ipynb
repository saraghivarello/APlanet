```python
import pandas as pd
import numpy as np
import math
```

## Import data

```python
file = r"C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_2\Data\Part2.csv"
df1 = pd.read_csv(file)

df1 = df1.dropna(subset=['player.Price_O_B'])

df1.drop(['participant._max_page_index','participant._current_app_name','participant._current_page_name', 'participant._index_in_pages',
          'participant._current_page_name','participant.time_started_utc','participant.visited'], inplace=True, axis=1)

df1 = df1.loc[:, (df1 != 0).any(axis=0)]
df1c = df1.dropna(how='all', axis=1)
lenght=len(df1c)
df1c.index  = np.arange(lenght)
```

## Add columns with text (of questions and answers)

```python
# module
v=[]
for i in df1['participant.id_in_session']:
    if math.isnan(i) is True:
        v.append('-')
    else: v.append((i-1)%5+1)
df1.insert(loc=1, column='mod', value=v)
```

```python
df1.head(6)
```

```python
n=5
for k in range(1,7):

    v=[]
    for i in df1['%d_policy_choice' %k]:
        if math.isnan(i) is True:
            v.append('-')
        elif i==1:
            v.append('A')
        elif i==2:
            v.append('B')
        elif i==3:
            v.append('None of them, keep the current system')
    df1.insert(loc=n, column='%d_What is your preferred policy type?' %k, value=v)
    n+=2
        
    for j in ['A','B']:
            
        v=[]
        for i in df1['%d_Price_UR_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.10 EUR')
            elif i==2:
                v.append('0.20 EUR')
            elif i==3:
                v.append('0.40 EUR')
            elif i==4:
                v.append('0.50 EUR')   
        df1.insert(loc=n, column='%d_%s_Price in urban areas, rush hour' %(k,j), value=v)
        n+=2

        # UN
        v=[]
        for i in df1['%d_Price_UN_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.10 EUR')
            elif i==2:
                v.append('0.20 EUR')
            elif i==3:
                v.append('0.40 EUR')
            elif i==4:
                v.append('0.50 EUR')   
        df1.insert(loc=n, column='%d_%s_Price in urban areas, outside rush hour' %(k,j), value=v)
        n+=2

        # O
        v=[]
        for i in df1['%d_Price_O_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.01 EUR')
            elif i==2:
                v.append('0.02 EUR')
            elif i==3:
                v.append('0.03 EUR')
            elif i==4:
                v.append('0.05 EUR')   
        df1.insert(loc=n, column='%d_%s_Price outside urban areas' %(k,j), value=v)
        n+=2
        
        # EV
        v=[]
        for i in df1['%d_EV_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.01 EUR')
            elif i==2:
                v.append('0.02 EUR')
            elif i==3:
                v.append('0.03 EUR')
            elif i==4:
                v.append('0.05 EUR')   
        df1.insert(loc=n, column='%d_%s_Price for electric vehicles' %(k,j), value=v)
        n+=2

        # Revenue
        v=[]
        for i in df1['%d_Revenue_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('General budget')
            elif i==2:
                v.append('Equal cash transfer')
            elif i==3:
                v.append('Cash transfer for low-income citizens')
            elif i==4:
                v.append('Investments in roads')
            elif i==5:
                v.append('Investments in public transport, walking and cycling')   
        df1.insert(loc=n, column='%d_%s_Use of revenue' %(k,j), value=v)
        n+=2
```
