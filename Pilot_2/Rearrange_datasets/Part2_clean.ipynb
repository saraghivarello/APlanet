```python
import pandas as pd
import numpy as np
import math
```

## Import data

```python
file = r"C:\Users\PcLaptop\Documents\GitHub\APlanet\Pilot_2\Rearrange_datasets\Part2_data_to_rows.csv"
df1 = pd.read_csv(file)

#df1 = df1.loc[:, (df1 != 0).any(axis=0)]
print(np.shape(df1))
df1 = df1.dropna(how='all', axis=1)
df1 = df1[df1['1_policy_choice'] != 0] #'q836peeq','yu68b1d9'
lenght=len(df1)
df1.index  = np.arange(lenght)
```

```python
df1.loc[:, (df1 != 0).any(axis=0)]
```

## Add columns with text (of questions and answers)

```python
df1.head(6)
```

```python
for k in range(1,9):

    v=[]
    for i in df1['%d_policy_choice' %k]:
        if math.isnan(i) is True:
            v.append('-')
        elif i==1:
            v.append('A')
        elif i==2:
            v.append('B')
        elif i==3:
            v.append('None of them, keep the current system')
    df1.insert(loc=df1.columns.get_loc('%d_policy_choice' %k)+1, column='%d_What is your preferred policy type?' %k, value=v)
        
    for j in ['A','B']:
            
        v=[]
        for i in df1['%d_Price_UR_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.10 EUR')
            elif i==2:
                v.append('0.20 EUR')
            elif i==3:
                v.append('0.40 EUR')
            elif i==4:
                v.append('0.50 EUR')   
        df1.insert(loc=df1.columns.get_loc('%d_Price_UR_%s' %(k,j))+1, column='%d_%s_Price in urban areas, rush hour' %(k,j), value=v)

        # UN
        v=[]
        for i in df1['%d_Price_UN_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.10 EUR')
            elif i==2:
                v.append('0.20 EUR')
            elif i==3:
                v.append('0.40 EUR')
            elif i==4:
                v.append('0.50 EUR')   
        df1.insert(loc=df1.columns.get_loc('%d_Price_UN_%s' %(k,j))+1, column='%d_%s_Price in urban areas, outside rush hour' %(k,j), value=v)

        # O
        v=[]
        for i in df1['%d_Price_O_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.01 EUR')
            elif i==2:
                v.append('0.02 EUR')
            elif i==3:
                v.append('0.03 EUR')
            elif i==4:
                v.append('0.05 EUR')   
        df1.insert(loc=df1.columns.get_loc('%d_Price_O_%s' %(k,j))+1, column='%d_%s_Price outside urban areas' %(k,j), value=v)

        
        # EV
        v=[]
        for i in df1['%d_EV_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('0.01 EUR')
            elif i==2:
                v.append('0.02 EUR')
            elif i==3:
                v.append('0.03 EUR')
            elif i==4:
                v.append('0.05 EUR')   
        df1.insert(loc=df1.columns.get_loc('%d_EV_%s' %(k,j))+1, column='%d_%s_Price for electric vehicles' %(k,j), value=v)


        # Revenue
        v=[]
        for i in df1['%d_Revenue_%s' %(k,j)]:
            if math.isnan(i) is True:
                v.append('-')
            elif i==1:
                v.append('General budget')
            elif i==2:
                v.append('Equal cash transfer')
            elif i==3:
                v.append('Cash transfer for low-income citizens')
            elif i==4:
                v.append('Investments in roads')
            elif i==5:
                v.append('Investments in public transport, walking and cycling')   
        df1.insert(loc=df1.columns.get_loc('%d_Revenue_%s' %(k,j))+1, column='%d_%s_Use of revenue' %(k,j), value=v)
```
